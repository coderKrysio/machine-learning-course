```python
from kmodes.kmodes import KModes
import numpy as np
```

```python
# Sample categorical data
data = np.array([
    ['red', 'SUV', 'domestic'],
    ['blue', 'sedan', 'domestic'],
    ['green', 'SUV', 'domestic'],
    ['red', 'sedan', 'domestic'],
    ['blue', 'SUV', 'import'],
    ['green', 'sedan', 'import'],
    ['red', 'SUV', 'import'],
    ['blue', 'sedan', 'import'],
])
data
```

```python
# Initialize the K-Modes algorithm
k_modes = KModes(n_clusters=2, init='Huang', n_init=5, verbose=1)
k_modes
```

```python
# Fit the model to the data
clusters = k_modes.fit_predict(data)
clusters
```

```python
# Print the cluster centroids
print('Cluster centroids:')
print(k_modes.cluster_centroids_)
```

```python
# Print the cluster assignments
print('Cluster assignments:')
print(clusters)
```

```python
import pandas as pd
from kmodes.kmodes import KModes
```

```python
# Load data from CSV file
data = pd.read_csv('homeprices.csv')
data
```

```python
# Convert the dataframe to a numpy array
data_array = data.to_numpy()
data_array
```

```python
# Initialize the K-Modes algorithm
k_modes = KModes(n_clusters=2, init='Huang', n_init=5, verbose=1)
k_modes
```

```python
# Fit the model to the data
clusters = k_modes.fit_predict(data_array)
clusters
```

```python
# Print the cluster centroids and the cluster assignments
print('Cluster centroids:')
print(k_modes.cluster_centroids_)
```

```python
# Print the cluster assignments
print('Cluster assignments:')
print(clusters)
```

```python
# Optionally, add the cluster assignments to the original dataframe
data['Cluster'] = clusters
data
```

```python
# Save the results to a new CSV file
data.to_csv('clustered_data.csv', index=False)
```
